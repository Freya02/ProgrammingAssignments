```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```
df=pd.read_csv("Topic_Survey_Assignment.csv",skiprows=1,names=['very_int','somewhat_int',"not_int",'total'],index_col=0,skipfooter=1)
```

```
df.head()
```

```
df.sort_values(by = ['very_int'],ascending=False,inplace=True)
df_survey_pct = ((df / 2233) * 100).round(2)
```

```
df_survey_pct.drop(labels='total',axis=1,inplace=True)
```

```

ax = df_survey_pct.plot(kind='bar',
                       figsize = (20, 8),
                       width = 0.8,
                       color = ['#5cb85c', '#5bc0de', '#d9534f'],
                       fontsize = 14)

plt.title('Percentage of Respondents\''' Interests in Data Science Areas', fontsize=16)
y_axis = ax.axes.get_yaxis()
y_axis.set_visible(False)
x_offset = -0.09
y_offset = 0.80
for p in ax.patches:
    b = p.get_bbox()
    #print(b)
    val = "{:.2%}".format((b.y1 + b.y0)*0.01)        
    ax.annotate(val, ((b.x0 + b.x1)/2 + x_offset, b.y1 + y_offset))

plt.show()

```

```
ax1 = df_survey_pct.plot.bar( figsize = (20, 8),
                       width = 0.8,color = ['#5cb85c', '#5bc0de', '#d9534f'])
for p in ax1.patches:
  ax1.annotate("{:.2%}".format(p.get_height()*0.01),
                xy=(p.get_x()+0.05, p.get_height()+1.04))
```
